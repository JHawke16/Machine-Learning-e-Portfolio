```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the Iris dataset
iris_path = 'C:/Users/Hari/Desktop/Uni Stuff/Module 5/e-Portfolio Submission/Unit06 iris.csv'
iris_data = pd.read_csv(iris_path)

print(iris_data.head())
```

```python
# Remove non-numeric columns
iris_data_numeric = iris_data.drop(columns=['species'])

print(iris_data_numeric.head())
```

```python
# Perform K-Means clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=0)
iris_data_numeric['Cluster'] = kmeans.fit_predict(iris_data_numeric)

iris_data_numeric['Cluster'] = iris_data_numeric['Cluster'].astype(str)

species_to_num = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
iris_data['species_num'] = iris_data['species'].map(species_to_num)

conf_matrix = confusion_matrix(iris_data['species_num'], iris_data_numeric['Cluster'].astype(int))

print(conf_matrix)

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Cluster 0', 'Cluster 1', 'Cluster 2'], yticklabels=['setosa', 'versicolor', 'virginica'])
plt.xlabel('Cluster')
plt.ylabel('Species')
plt.title('Confusion Matrix: Iris Dataset Clusters vs Species')
plt.show()
```

```python
# Loading the Wine dataset
wine_path = 'C:/Users/Hari/Desktop/Uni Stuff/Module 5/e-Portfolio Submission/Unit06 wine.csv'
wine_data = pd.read_csv(wine_path)

print(wine_data.head())
```

```python
# Remove non-numeric columns
wine_data_numeric = wine_data.drop(columns=['Wine'])

print(wine_data_numeric.head())
```

```python
# Perform K-Means clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=0)
wine_data_numeric['Cluster'] = kmeans.fit_predict(wine_data_numeric)

wine_data_numeric['Cluster'] = wine_data_numeric['Cluster'].astype(str)

wine_labels_to_num = {1: 0, 2: 1, 3: 2}

wine_data['wine_num'] = wine_data['Wine'].map(wine_labels_to_num)

conf_matrix = confusion_matrix(wine_data['wine_num'], wine_data_numeric['Cluster'].astype(int))

print(conf_matrix)

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Cluster 0', 'Cluster 1', 'Cluster 2'], yticklabels=[1, 2, 3])
plt.xlabel('Cluster')
plt.ylabel('Wine Label')
plt.title('Confusion Matrix: Wine Dataset Clusters vs Wine Labels')
plt.show()
```

```python
# Loading the WeatherAUS dataset
weather_path = 'C:/Users/Hari/Desktop/Uni Stuff/Module 5/e-Portfolio Submission/Unit06 weatherAUS.csv'
weather_data = pd.read_csv(weather_path)

print(weather_data.head())
```

```python
# Remove non-numeric columns and fill missing values
weather_data_numeric = weather_data.select_dtypes(include=[float, int]).dropna()

print(weather_data_numeric.head())
```

```python
# Perform K-Means clustering with K values ranging from 2 to 6
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=0)
    weather_data_numeric[f'Cluster_{k}'] = kmeans.fit_predict(weather_data_numeric)
    
    # Visualize the clusters
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=weather_data_numeric, x=weather_data_numeric.columns[0], y=weather_data_numeric.columns[1], hue=f'Cluster_{k}', palette='viridis')
    plt.title(f'K-Means Clustering with K={k} on WeatherAUS Data')
    plt.show()
```
